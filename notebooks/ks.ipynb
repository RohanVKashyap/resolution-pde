```python
import torch
import matplotlib.pyplot as plt
```

```python
data = torch.load('/home/rvk/T=20,niu=0.01,N=1024,dt=0.001,6pi,dtsave=0.1,sample=51_ut.pt', map_location=torch.device('cpu'))
print(data.shape)
```

```python
import matplotlib.pyplot as plt

total = 20
ncols = 5
nrows = total // ncols

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 6))
axes = axes.flatten()

for i in range(total):
    axes[i].plot(data[0][i].numpy(), c='r')
    axes[i].set_title(f'Time {i}')

plt.tight_layout()
plt.show() 
```
